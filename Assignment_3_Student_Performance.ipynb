```python
"""Qn3:-3. Who scores the most on average for math, reading and writing based on 
● Gender 
● Test preparation course"""
```

```python
df1 = df.groupby(['gender','test preparation course'])[['math score','reading score','writing score']].mean()
df1.sort_values('math score', ascending = False)
```

```python
#Conclusion 3:-
#Male students who completed  test prepration is have most average in math score
#Female students who completed  test prepration is have most average in reading and writing score
```

```python
"""Qn4:- What do you think about the scoring variation for math, reading and writing based on ● Gender ● Test preparation course"""
```

```python
df2 = df.groupby(['gender','test preparation course'])[['math score','reading score','writing score']].var()
df2.sort_values('math score', ascending = False)
```

```python
###Conclusion4:- 
### Female students who doesn't attended test prepration have high variance in scoring, that means their score is 
###spreaded away from the mean value they need to attended the test preration course
```
